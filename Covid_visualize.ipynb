```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
covid_data = pd.read_csv('covid_cleaned.csv')
print(covid_data)
```

```python
black = covid_data[covid_data['race_ethnicity_combined'] == 'Black, Non-Hispanic']
```

```python
white = covid_data[covid_data['race_ethnicity_combined'] == 'White, Non-Hispanic']
```

```python
hispanic = covid_data[covid_data['race_ethnicity_combined'] == 'Hispanic/Latino']
```

```python
other = covid_data[covid_data['race_ethnicity_combined'] == 'Multiple/Other, Non-Hispanic']
```

```python
pacific_islander = covid_data[covid_data['race_ethnicity_combined'] == 'Native Hawaiian/Other Pacific Islander, Non-Hispanic']
```

```python
asian = covid_data[covid_data['race_ethnicity_combined'] == 'Asian, Non-Hispanic']
```

```python
american_indian = covid_data[covid_data['race_ethnicity_combined'] == 'American Indian/Alaska Native, Non-Hispanic']
```

```python
df = pd.DataFrame({'Race/Ethnicity': ['Black','White','Hispanic','Other','Pacific Islander','Asian','American Indian'],'Num of People': [108701,748270, 174870, 23176, 6065,28638,4831]})
```

```python
df.groupby(['Race/Ethnicity']).sum().plot(kind='pie', y='Num of People', autopct='%1.0f%%')
```

```python
explode = (0, 0, 0, 0.7,0.7,0.7,0.7)
sizes = [108701,748270, 174870, 23176, 6065,28638,4831]
labels = 'Black','White','Hispanic','Other','Pacific Islander','Asian','American Indian'



fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels,autopct='%1.1f%%',
        shadow=True)
plt.show()
```

```python
ax = black.groupby('age_group')['sex'].value_counts().unstack().sort_index(ascending=True).plot(kind='bar', stacked = True)
ax.set_xlabel("Age Groups")
ax.set_ylabel("Frequency")
ax.set_title('Black')
```

```python
ax = white.groupby('age_group')['sex'].value_counts().unstack().sort_index(ascending=True).plot(kind='bar', stacked = True)
ax.set_xlabel("Age Groups")
ax.set_ylabel("Frequency")
ax.set_title('White')
```

```python
ax = hispanic.groupby('age_group')['sex'].value_counts().unstack().sort_index(ascending=True).plot(kind='bar', stacked = True)
ax.set_xlabel("Age Groups")
ax.set_ylabel("Frequency")
ax.set_title('Hispanic')
```

```python
ax = other.groupby('age_group')['sex'].value_counts().unstack().sort_index(ascending=True).plot(kind='bar', stacked = True)
ax.set_xlabel("Age Groups")
ax.set_ylabel("Frequency")
ax.set_title('Other')
```

```python
ax = asian.groupby('age_group')['sex'].value_counts().unstack().sort_index(ascending=True).plot(kind='bar', stacked = True)
ax.set_xlabel("Age Groups")
ax.set_ylabel("Frequency")
ax.set_title('Asian')
```

```python
ax = pacific_islander.groupby('age_group')['sex'].value_counts().unstack().sort_index(ascending=True).plot(kind='bar', stacked = True)
ax.set_xlabel("Age Groups")
ax.set_ylabel("Frequency")
ax.set_title('Pacific Islander')
```

```python
ax = american_indian.groupby('age_group')['sex'].value_counts().unstack().sort_index(ascending=True).plot(kind='bar', stacked = True)
ax.set_xlabel("Age Groups")
ax.set_ylabel("Frequency")
ax.set_title('American Indian')
```
